Babar Carrer Overview

```python
df['Runs'].describe()
```

```python
totalRuns = df['Runs'].sum()
totalBallFaced = df['BF'].sum()
totalInnings = df['Innings'].count()
totalAverage = (df['Runs'].sum() /(~df['Dismissal'].str.contains('not out', case=False)).sum()).round(2)
strikeRate = (totalRuns / totalBallFaced) * 100
totalBoundries = df['4s'].sum() + df['6s'].sum()
totalFours = df['4s'].sum()
totalSixes = df['6s'].sum()
totalNotout = df['Dismissal'].str.contains('not out').sum()
highestScore = df['Runs'].agg('max')
totalCenturies = (df['Runs'] >= 100).sum()
totalFifty = ((df['Runs'] >= 50) & (df['Runs']<100)).sum()
totalDucks = (df['Runs'] == 0).sum()
print(f"Total Innings: {totalInnings}")
print(f"Total Not Out: {totalNotout}")
print(f"Total Score: {totalRuns}")
print(f"Highest Score: {highestScore}")
print(f"Average Score: {totalAverage}")
print(f"Total Ball Faced: {totalBallFaced}")
print(f"Strike Rate: {strikeRate.round(2)}")
print(f"Total Centuries (100s): {totalCenturies}")
print(f"Total Fifties (50s): {totalFifty}")
print(f"Total 4s: {totalFours}")
print(f"Total 6s: {totalSixes}")
print(f"Total Ducks (0s): {totalDucks}")
print(f"Total Boundries: {totalBoundries}")
```

Babar Scores Against Opposition Q: (Is Babar a Zimbabar or Zimbu?)

```python
df.groupby('Opposition').agg(
    Matches=('Runs', 'count'),
    TotalRuns=('Runs', 'sum'),
    StrikeRate=('Runs', lambda x: (x.sum() / df.loc[x.index, 'BF'].sum()) * 100),
    BallFaced = ('BF','sum'),
    HighestScore=('Runs', 'max'),
    BattingAverage=('Runs', lambda x: x.sum() / (~df.loc[x.index, 'Dismissal'].str.contains('not out')).sum())
).sort_values(by='TotalRuns', ascending=False).round(2)
```

Babar Score Per Calender Year (Q: In Which Year was babar dominated?)

Visualizations:

```python
yearlyRuns = df.groupby('Year')['Runs'].sum()

plt.figure(figsize=(6,4))
plt.bar(yearlyRuns.index, yearlyRuns.values, color='skyblue')
plt.plot(yearlyRuns.index, yearlyRuns.values, color='red')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Runs', fontsize=12)
plt.title('Babar Azam (Year-wise Total Runs)', fontsize=14, fontweight='bold')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
yearlyAvg = df.groupby('Year').apply(
    lambda x: x['Runs'].sum() / (~x['Dismissal'].str.contains('not out', case=False)).sum()
    , include_groups=False
)

plt.figure(figsize=(8,3))
plt.plot(yearlyAvg.index, yearlyAvg.values, marker='o', linestyle='-', color='green')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Batting Average', fontsize=12)
plt.title('Babar Azam (Year-wise Batting Average)', fontsize=14, fontweight='bold')
plt.xticks(rotation=0)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
oppRuns = df.groupby('Opposition')['Runs'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,4))
plt.bar(oppRuns.index, oppRuns.values, color='orange')
plt.xlabel('Opposition', fontsize=12)
plt.ylabel('Total Runs', fontsize=12)
plt.title('Babar Azam (Runs against Opposition)', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
homeawayStats = df.groupby('HomeAway').agg(
    Average=('Runs', lambda x: x.sum()/ (~df.loc[x.index,'Dismissal'].str.contains('not out',case=False)).sum()),
    StrikeRate=('Runs', lambda x: (x.sum() / df.loc[x.index,'BF'].sum())*100)
)

x = np.arange(len(homeawayStats))
width = 0.35

plt.figure(figsize=(6,3))
plt.bar(x - width/2, homeawayStats['Average'], width, label='Average')
plt.bar(x + width/2, homeawayStats['StrikeRate'], width, label='StrikeRate')

plt.xlabel('Venue Type', fontsize=12)
plt.ylabel('Runs / Strike Rate', fontsize=12)
plt.title('Babar Azam - (Home/Away Performance)', fontsize=14, fontweight='bold')
plt.xticks(x, homeawayStats.index, rotation=45)
plt.legend()
plt.show()
```

```python
dismissals = df[~df['Dismissal'].str.contains('not out', case=False)]['Dismissal'].value_counts()

plt.figure(figsize=(6,4))
plt.barh(dismissals.index, dismissals.values, color='red')
plt.xlabel('Count', fontsize=12)
plt.ylabel('Dismissal Type', fontsize=12)
plt.title('Babar Azam - Dismissal Types', fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
battingPos = df.groupby('Pos')['Runs'].agg(
    'sum'
).sort_values(ascending=False)

plt.figure(figsize=(6,3))
plt.bar(battingPos.index.astype(str), battingPos.values, color='purple', alpha=0.7)
plt.xlabel('Batting Position', fontsize=12)
plt.ylabel('Total Runs', fontsize=12)
plt.title('Babar Azam (Runs by Batting Position)', fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
boundaries = df.groupby('Opposition')[['4s','6s']].sum().sort_values('6s', ascending=False)
width = 0.35
x = np.arange(len(boundaries))

plt.figure(figsize=(8,4))
plt.bar(x - width/2, boundaries['4s'], width, label='Fours', alpha=0.7)
plt.bar(x + width/2, boundaries['6s'], width, label='Sixes', alpha=0.7)
plt.xlabel('Opposition', fontsize=12)
plt.ylabel('Boundaries', fontsize=12)
plt.title('Babar Azam(Fours and Sixes vs Opposition)', fontsize=14, fontweight='bold')
plt.xticks(x, boundaries.index, rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
topSR = df.groupby('Opposition').agg(
    StrikeRate=('Runs', lambda x: (x.sum() / df.loc[x.index,'BF'].sum())*100)
).sort_values(by='StrikeRate', ascending=False).head(5)

plt.figure(figsize=(12,3))
plt.bar(topSR.index, topSR['StrikeRate'], color='teal')
plt.xlabel('Opposition', fontsize=12)
plt.ylabel('Strike Rate', fontsize=12)
plt.xticks(rotation=45)
plt.title('Babar Azam (Top 5 Strike Rate vs Opposition)', fontsize=14, fontweight='bold')
plt.show()
```
